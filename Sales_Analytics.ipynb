```python
# Year from the Date

retail_bk['year1'] = pd.DatetimeIndex(retail_bk['Date']).year

# Month from the Date

retail_bk['month1'] = pd.DatetimeIndex(retail_bk['Date']).month

# Day from the Date

retail_bk['Day1'] = pd.DatetimeIndex(retail_bk['Date']).day

# Display first five records

retail_bk.head()
```
